```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ti = sns.load_dataset('titanic')
ti
```

```python
sns.relplot(data = ti, x = 'age', y = "fare")
```

```python
sns.relplot(data = ti, x = 'age', y = "fare", hue = "alive")
```

```python
sns.relplot(data = ti, x = 'age', y = "fare", hue = "alive",
           col = "pclass", height = 4, aspect = 0.8)
```

```python
sns.relplot(data = ti, x = 'age', y = "fare", hue = "alive",
           col = "pclass", row = "embark_town",  height = 4, aspect = 0.8)
```

```python
sns.relplot(data = ti, x = 'age', y = "fare", hue = "alive",
           col = "pclass", height = 4, aspect = 0.8, style = "sex")
```

```python
sns.relplot(data = ti, x = 'age', y = "fare", hue = "alive",
           col = "pclass", height = 4, aspect = 0.8, style = "sex", kind = "line")
```

```python
sns.relplot(data = ti, x = 'age', y = "fare", kind = 'line')
```

```python
sns.relplot(data = ti, x = 'age', y = "fare", kind = 'line', ci = None, hue = "sex", 
            dashes = True, style = "sex", markers= True)
```

```python
sns.relplot(data = ti, x = 'age', y = "fare", kind = 'line', ci = None, hue = "sex", 
            dashes = True, style = "sex", markers= True)
```
