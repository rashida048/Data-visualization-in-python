```python
from mpl_toolkits.axes_grid1.inset_locator import mark_inset, inset_axes

#plt.rcParams["figure.figsize"] = [8, 5]
#plt.rcParams["figure.autolayout"] = True

plt.figure(figsize = (8, 5))
x = d['length']
y = d['price']

ax = plt.subplot(1, 1, 1)
ax.scatter(x, y)
ax.set_xlabel("Length")
ax.set_ylabel("Price")

axins = inset_axes(ax, 1, 1, loc = 1, bbox_to_anchor=(0.3, 0.7),
                   bbox_transform = ax.figure.transFigure)

axins.scatter(x, y)

x1, x2 = 0.822, 0.838
y1, y2 = 6400, 12000

axins.set_xlim(x1, x2)
axins.set_ylim(y1, y2)

mark_inset(ax, axins, loc1=1, loc2=3, fc="none", ec = "0.4")
plt.show()
```

```python
from matplotlib.patches import ConnectionPatch

fig = plt.figure(figsize=(6, 5))
plt.subplots_adjust(bottom = 0., left = 0, top = 1., right = 1)

sub1 = fig.add_subplot(2,2,1) # two rows, two columns, fist cell
sub1.scatter(d['length'], d['width'], color = 'red')
sub1.set_xlim(0.80, 0.85)
sub1.set_ylim(0.88, 0.91)
sub1.set_ylabel('width', labelpad = 5)

sub2 = fig.add_subplot(2, 2, 2)
sub2.scatter(d['length'], d['width'], color = 'blue')
sub2.set_xlim(0.85, 0.90)
sub2.set_ylim(.92, 0.96)

sub3 = fig.add_subplot(2,2,(3,4)) # two rows, two colums, combined third and fourth cell
sub3.scatter(d['length'], d['width'], color = 'darkorchid', alpha = .7)
#sub3.set_xlim(0.80, 0.85)
#sub3.set_ylim(0.88, 0.91)
sub3.set_xlabel(r'$\theta$ (rad)', labelpad = 5)
sub3.set_ylabel('y', labelpad = 5)

sub3.fill_between((0.80, 0.85), 0.8, 1, facecolor= "red", alpha = 0.2)
sub3.fill_between((0.85, 0.90), 0.8, 1, facecolor= "blue", alpha = 0.2)

con1 = ConnectionPatch(xyA = (0.8, 0.88), coordsA=sub1.transData,
                       xyB=(0.80, 0.91), coordsB=sub3.transData, color = 'red')
fig.add_artist(con1)

con2 = ConnectionPatch(xyA = (0.85, 0.88), coordsA=sub1.transData,
                       xyB=(0.85, 0.91), coordsB=sub3.transData, color = 'red')
fig.add_artist(con2)

con3 = ConnectionPatch(xyA = (0.85, 0.92), coordsA=sub2.transData,
                       xyB=(0.85, 0.91), coordsB=sub3.transData, color = 'blue')
fig.add_artist(con3)

con4 = ConnectionPatch(xyA = (0.90, 0.92), coordsA=sub2.transData,
                       xyB=(0.90, 0.95), coordsB=sub3.transData, color = 'blue')
fig.add_artist(con4)
```

```python
from matplotlib.patches import ConnectionPatch

fig = plt.figure(figsize=(8, 5))
#plt.subplots_adjust(bottom = 0., left = 0, top = 0, right = 1)

plot1 = fig.add_subplot(2,2,1) # two rows, two columns, fist cell
plot1.scatter(d['city-mpg'], d['highway-mpg'], color = 'red')
plot1.set_xlim(15, 20)
plot1.set_ylim(17, 25)
plot1.set_ylabel('highway-mpg', labelpad = 5)

plot2 = fig.add_subplot(2, 2, 2)
plot2.scatter(d['city-mpg'], d['highway-mpg'], color = 'blue')
plot2.set_xlim(25, 30)
plot2.set_ylim(25, 35)

plot3 = fig.add_subplot(2,2,(3,4)) # two rows, two colums, combined third and fourth cell
plot3.scatter(d['city-mpg'], d['highway-mpg'], color = 'darkorchid', alpha = .7)
plot3.set_xlabel('city-mpg', labelpad = 5)
plot3.set_ylabel('highway-mpg', labelpad = 5)

plot3.fill_between((15, 20), 10, 60, facecolor= "red", alpha = 0.2)
plot3.fill_between((25, 30), 10, 60, facecolor= "blue", alpha = 0.2)


conn1 = ConnectionPatch(xyA = (15, 17), coordsA=plot1.transData,
                       xyB=(15, 20), coordsB=plot3.transData, color = 'red')
fig.add_artist(conn1)

conn2 = ConnectionPatch(xyA = (20, 17), coordsA=plot1.transData,
                       xyB=(20, 20), coordsB=plot3.transData, color = 'red')
fig.add_artist(conn2)

conn3 = ConnectionPatch(xyA = (25, 25), coordsA=plot2.transData,
                       xyB=(25, 30), coordsB=plot3.transData, color = 'blue')
fig.add_artist(conn3)

conn4 = ConnectionPatch(xyA = (30, 25), coordsA=plot2.transData,
                       xyB=(30, 30), coordsB=plot3.transData, color = 'blue')
fig.add_artist(conn4)
```
